```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1,inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

#Analysis

<h2>Gender</h2>

```python
#Gender vs Count
plt.figure(figsize=(5,5))
gender_cnt_plot = sns.countplot(data = df,x = "Gender",palette=["Green","Red"])
gender_cnt_plot

for cnt in gender_cnt_plot.containers:
  gender_cnt_plot.bar_label(cnt)
```

```python
#Gender vs Total Amount

plt.figure(figsize=(5,5))
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_gen,x = "Gender",y = "Amount",palette=['red','black'])
```

<h2>Age</h2>

```python
plt.figure(figsize=(13,5))

age_cnt_plot = sns.countplot(data = df,x = "Age Group",hue = "Gender",palette = ["Green","Red"])

for cnt in age_cnt_plot.containers:
  age_cnt_plot.bar_label(cnt)
```

```python
#Total Amount vs Age Group

sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = "Amount",ascending = False)
sns.barplot(x= "Age Group",y = "Amount",data = sales_age,palette=['Green','orange','yellow','grey','pink','black','brown'])
```

<h2>State</h2>

```python
#total order from 10 states

sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (17,5)})
sns.barplot(data = sales_state,x = 'State',y = 'Orders',hue = 'State')
```

<h2>Marital Status</h2>

```python
plt.figure(figsize = (5,5))
maritial_status_cnt_plot = sns.countplot(data = df,x = 'Marital_Status',palette = ['red',"black"])
# sns.set(rc = {'figure.figsize':(4,5)})
for cnt in maritial_status_cnt_plot.containers:
  maritial_status_cnt_plot.bar_label(cnt)
```

```python
#Gender vs maritial_Status
maritial_status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc = {'figure.figsize' :(6,5)})
sns.barplot(data = maritial_status,x = 'Marital_Status',y = 'Amount',hue = "Gender",palette = ['blue','black'])
```

<h2>Occupation</h2>


```python
sns.set(rc = {'figure.figsize' : (20,6)})
occupation_cnt = sns.countplot(data = df,x = 'Occupation')


for cnt in occupation_cnt.containers:
  occupation_cnt.bar_label(cnt)
```

```python
plt.figure(figsize = (7,5))
occupation_wise_amount = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = "Amount",ascending = False).head(5)
# sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = occupation_wise_amount,hue= 'Occupation',y = 'Amount')
```

<h2>Product Category</h2>

```python
sns.set(rc = {'figure.figsize':(25,6)})
product_category_count = sns.countplot(data = df,x = 'Product_Category')

for cnt in product_category_count.containers:
  product_category_count.bar_label(cnt)
```

```python
sales_of_product_category = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_of_product_category,x = 'Product_Category',y = 'Amount')
```

```python
sales_most_product = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = "Orders",ascending = False).head(10)
plt.figure(figsize=(14,5))
sns.barplot(data = sales_most_product,x = "Product_ID",y = "Orders")
```
